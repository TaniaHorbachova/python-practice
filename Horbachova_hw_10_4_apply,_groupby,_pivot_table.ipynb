4. –Ø–∫–∞ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è —Ç–æ–≤–∞—Ä—ñ–≤ –º–∞—î –Ω–∞–π–Ω–∏–∂—á—ñ –ø—Ä–æ–¥–∞–∂—ñ —É –∫–æ–∂–Ω–æ–º—É –º—ñ—Å—Ç—ñ?

   –ì—ñ–ø–æ—Ç–µ–∑–∞: —É —Ä—ñ–∑–Ω–∏—Ö –º—ñ—Å—Ç–∞—Ö —î –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó, —è–∫—ñ –º–∞–π–∂–µ –Ω–µ –∫—É–ø—É—é—Ç—å.

```python
invoices_by_category = df.pivot_table(index='City', columns='Product line', values='Invoice ID', aggfunc='count')
invoices_by_category
```

```python
def min_categories(row):
    min_val = row.min()
    return list(row[row == min_val].index)

min_cats = invoices_by_category.apply(min_categories, axis=1)

min_vals = invoices_by_category.min(axis=1)

result = pd.DataFrame({'City': min_cats.index, 'MinCategories': min_cats.values, 'MinSales': min_vals.values})
display(result.style.hide(axis="index"))
```
